2. Pesquise a função "explode" no pandas. Ela pode ser útil.

```python
import numpy as np
import pandas as pd
```

```python
df_movies = pd.read_csv('dados/movies.csv', sep=',')
df_movies.head()
```

```python
df_movies.info()
```

```python
df_movies['genres_count'] = df_movies['genres'].str.split("|").str.len()
df_movies['generos'] = df_movies['genres'].str.split("|")
df_movies = df_movies.explode('generos')
df_movies
```

```python
df_ratings = pd.read_csv('dados/ratings.csv', sep=',')
df_ratings.head()
```

```python
df_ratings.info()
```

```python
df_movies_ratings = pd.merge(df_movies, df_ratings, on='movieId')
df_movies_ratings
```

### Quantos gêneros o filme "The Matrix" tem? Quais são os gêneros?

```python
df_movies[df_movies['title'].str.contains('Matrix')]
```

```python
qtde_generos = df_movies['genres_count'][df_movies['title'].str.contains('Matrix, The')]
qtde_generos.unique()
```

```python
generos = df_movies['generos'][df_movies['title'].str.contains('Matrix, The')]
generos
```

### Quantos gêneros cada filme tem, em média?

```python
df_movies['genres_count'].mean()
```

### Quais são os cinco filmes com maior número de gêneros?

```python
df_movies.sort_values(by='genres_count', ascending=False).head(5)
```

### Quantos usuários deram nota para o filme "The Matrix"?

```python
df_movies_ratings[df_movies_ratings['title'].str.contains('Matrix, The')]['userId'].count()
```

### Quantos usuários, em média, deram nota para cada filme?

```python
df_ratings_user = df_movies_ratings.groupby('movieId')['userId'].count().reset_index()
df_ratings_user
```

```python
df_ratings_user['userId'].mean()
```

### Quais são os cinco filmes para os quais mais usuários deram nota?

```python
df_ratings_user.sort_values(by='userId', ascending=False).head(5)
```

### Qual o rating médio do filme "The Matrix"?

```python
df_movies_ratings[df_movies_ratings['title'].str.contains('Matrix, The')]['rating'].mean()
```

### Cada filme tem um rating médio. Qual a média desses ratings, por gênero?

```python
df_movies['generos'].nunique()
```

```python
df_rating_mean = df_movies_ratings.groupby('movieId')['rating'].mean().reset_index().round(2)
df_rating_mean
```

```python
df_movies_v2 = pd.merge(df_movies, df_rating_mean, on='movieId')
df_movies_v2
```

```python
df_movies_v2.groupby('generos')['rating'].mean().reset_index()
```

### Para cada gênero de filme, quantos usuários deram cada nota possível?

```python
df_movies_ratings.groupby(['generos', 'rating']).agg({'userId':'count'})
```
